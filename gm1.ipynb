# TP : Simulation de File d'Attente G/M/1 avec SimPy

Ce notebook simule une **file d’attente G/M/1** (temps d'arrivée **général** et service **exponentiel**) en utilisant **Python** et la bibliothèque **SimPy**.

---

## Objectifs

- Simuler **au moins 1 000 000** de clients par expérience
- Faire varier le **taux d’arrivée** λ de **0.1 à 0.9**
- Répéter chaque expérience pour **obtenir des résultats stables**
- Calculer les mesures de performance suivantes :
  - **Temps de réponse moyen**  
  - **Temps d’attente moyen en file**  
  - **Taux d’occupation du serveur** ρ (rho), défini comme :   ρ = λ / μ

---

## Paramètres du Modèle

| Paramètre      | Valeur              |
|----------------|---------------------|
| Taux de service μ | 1.0                 |
| Nombre de clients | 1 000 000            |
| Répétitions par λ | 5                   |
| Lois de temps     | Arrivées : **uniforme**<br>Services : **exponentielle** |

---

## Mesures Visualisées

Le notebook produit des **graphiques** montrant l’évolution des mesures en fonction du taux d’arrivée λ :

1. **Temps de réponse moyen vs λ**
2. **Temps d’attente moyen vs λ**
3. **Utilisation du serveur (ρ) vs λ**

---

## Bonus : Analyse des temps entre arrivées

Une **distribution des temps inter-arrivées** est aussi représentée par un histogramme pour illustrer la loi uniforme.

---

> **Conclusion attendue :** Plus λ augmente, plus le système est sollicité, ce qui allonge les temps d'attente et de réponse, et pousse le serveur à saturation (ρ → 1).


```python
import simpy
import random
import numpy as np
import matplotlib.pyplot as plt

# Paramètres
MU = 1.0
NUM_CLIENTS = 1000000
REPEATS = 5
LAMBDA_VALUES = np.arange(0.1, 1.0, 0.1)

# Résultats à stocker
gm1_results = {
    'lambda': [],
    'response_times': [],
    'waiting_times': [],
    'utilization': []
}

# Classe G/M/1 avec temps d’arrivée uniforme
class GM1Queue:
    def __init__(self, env, arrival_rate, service_rate):
        self.env = env
        self.server = simpy.Resource(env, capacity=1)
        self.arrival_rate = arrival_rate
        self.service_rate = service_rate
        self.wait_times = []
        self.response_times = []
        self.busy_time = 0
        self.client_count = 0
        self.arrival_times = []  # Stocke les temps d'arrivée

    def arrival_process(self):
        mean_inter = 1 / self.arrival_rate
        a, b = 0.5 * mean_inter, 1.5 * mean_inter  # Uniforme autour de la moyenne
        while self.client_count < NUM_CLIENTS:
            inter_arrival = random.uniform(a, b)
            yield self.env.timeout(inter_arrival)
            self.client_count += 1
            self.arrival_times.append(self.env.now)  # Enregistre le temps d'arrivée
            self.env.process(self.service_process())

    def service_process(self):
        arrival_time = self.env.now
        with self.server.request() as req:
            yield req
            wait = self.env.now - arrival_time
            self.wait_times.append(wait)

            service_time = random.expovariate(self.service_rate)
            self.busy_time += service_time
            yield self.env.timeout(service_time)

            response_time = self.env.now - arrival_time
            self.response_times.append(response_time)

# Simulation pour chaque λ
last_queue = None  # Pour stocker la dernière file simulée

for lambd in LAMBDA_VALUES:
    gm1_resp, gm1_wait, gm1_util = [], [], []

    for _ in range(REPEATS):
        random.seed()
        env = simpy.Environment()
        queue = GM1Queue(env, lambd, MU)
        env.process(queue.arrival_process())
        env.run()

        gm1_resp.append(np.mean(queue.response_times))
        gm1_wait.append(np.mean(queue.wait_times))
        gm1_util.append(queue.busy_time / env.now)

        last_queue = queue  # Stocke la dernière queue

    gm1_results['lambda'].append(lambd)
    gm1_results['response_times'].append(np.mean(gm1_resp))
    gm1_results['waiting_times'].append(np.mean(gm1_wait))
    gm1_results['utilization'].append(np.mean(gm1_util))
```

```python
# Tracer les résultats principaux
plt.figure(figsize=(15, 12))

plt.subplot(3,1,1)
plt.plot(gm1_results['lambda'], gm1_results['response_times'], label='G/M/1 (uniform)', marker='x', color='blue')
plt.title('Temps de réponse moyen vs λ (G/M/1)')
plt.xlabel('λ (taux d\'arrivée)')
plt.ylabel('Temps de réponse moyen')
plt.grid(True)

plt.subplot(3,1,2)
plt.plot(gm1_results['lambda'], gm1_results['waiting_times'], label='G/M/1 (uniform)', marker='x', color='green')
plt.title('Temps d\'attente moyen en file vs λ (G/M/1)')
plt.xlabel('λ (taux d\'arrivée)')
plt.ylabel('Temps d\'attente moyen')
plt.grid(True)

plt.subplot(3,1,3)
plt.plot(gm1_results['lambda'], gm1_results['utilization'], label='G/M/1 (uniform)', marker='x', color='red')
plt.title('Utilisation du serveur vs λ (G/M/1)')
plt.xlabel('λ (taux d\'arrivée)')
plt.ylabel('Utilisation (ρ)')
plt.grid(True)

plt.tight_layout()
plt.show()
```

```python
# Visualisation des temps entre arrivées (pour la dernière simulation)
inter_arrival_times = np.diff(last_queue.arrival_times)
plt.figure(figsize=(10, 4))
plt.hist(inter_arrival_times, bins=50, color='skyblue', edgecolor='black')
plt.title('Distribution des temps entre arrivées (λ = {:.1f})'.format(last_queue.arrival_rate))
plt.xlabel('Temps entre deux arrivées')
plt.ylabel('Fréquence')
plt.grid(True)
plt.show()
```
